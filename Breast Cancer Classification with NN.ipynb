```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn.datasets
from sklearn.model_selection import train_test_split
```

```python
breast_cancer_dataset = sklearn.datasets.load_breast_cancer()
```

```python
breast_cancer_dataframe = pd.DataFrame(breast_cancer_dataset.data, columns = breast_cancer_dataset.feature_names)
```

```python
breast_cancer_dataframe['label'] = breast_cancer_dataset.target
```

```python
X = breast_cancer_dataframe.drop(columns = 'label', axis=1)
Y = breast_cancer_dataframe['label']
```

```python
print(X.shape, Y.shape)
```

# Train Test Split 

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=1)
```

```python
print(X_train.shape, Y_train.shape,X_test.shape, Y_test.shape)
```

# Standardize Data

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
X_train_std = scaler.fit_transform(X_train)
X_test_std = scaler.transform(X_test)
```

# Building a Neural Network

```python
import tensorflow as tf
tf.random.set_seed(3)
from tensorflow import keras
```

```python
#setting up layers of neural networks

model = keras.Sequential([
    keras.layers.Flatten(input_shape=(30,)),
    keras.layers.Dense(10, activation ='relu'),
    keras.layers.Dense(2, activation = 'sigmoid')
])
```

```python
model.compile(optimizer = 'adam',
             loss = 'sparse_categorical_crossentropy',
             metrics = ['accuracy'])
```

```python
history = model.fit(X_train, Y_train, validation_split = 0.1, epochs =20)
```

# Visualizing accuracy and loss

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])

plt.title('model_accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')

plt.legend(['training data', 'validation data'], loc = 'lower right')
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])

plt.title('model_loss')
plt.ylabel('loss')
plt.xlabel('epoch')

plt.legend(['training data', 'validation data'], loc = 'upper right')
```
