```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import StandardScaler
```

```python
data = pd.read_csv('data.csv')
```

```python
data
```

```python
data.describe()
```

```python
data['diagnosis'].value_counts()
```

```python
X = data.drop(columns = {'diagnosis','Unnamed: 32'}, axis=1)
Y = data['diagnosis']
```

```python
print(X.shape, Y.shape)
```

```python
Scaler = StandardScaler()
Scaler.fit(X)
standardized_data = Scaler.transform(X)
```

```python
print(standardized_data)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(standardized_data, Y, test_size=0.2, stratify=Y, random_state=1)
```

```python
print(X_train.shape, X_test.shape, Y_train.shape, Y_test.shape)
```

# Training the model

```python
model = LogisticRegression()
```

```python
model.fit(X_train, Y_train)
```

```python
training_data_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(training_data_prediction, Y_train)
print(training_data_accuracy*100)
```

```python
test_data_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(test_data_prediction, Y_test)
print(test_data_accuracy*100)
```
